```python
import matplotlib.pyplot as plt
import numpy as np

X_val = np.arange(1, 10001)
```

```python
#02:4A ESP32

print("Angle between the lines:", str(round(np.degrees(np.arctan(3.4028) - np.arctan(1.2137)),2)) + "°")

plt.plot(X_val, -490.1667 + 1.2137*X_val, color = 'green', linewidth=2, label = "Ground Truth")
plt.plot(X_val, -182.1706 + 3.4028*X_val, color = 'red', linewidth=2, label = "Total Drift")
plt.xlabel("Reference Time (in seconds)")
plt.ylabel("Time Difference")
plt.legend()
plt.show()
```
